```python
#La fabrica, juego
print("Bienvenido a la fabrica!")
print("en este juego debes avanzar por un tablero de casillas, el primero en llegar a la casilla final, logra restablecer el sistema de la fabrica y ganar la partida.")
print("reglas: \n El juego inicia en la casilla INICIO \n Por cada turno, un jugador lanza un dado y avanza ese numero de casillas \n Si un jugador cae en una casilla espicial, activa su efecto \n El jugador que llegue exactamente a la casilla LLEGADA, gana")
print("\nBienvenidos jugadores 1 y 2")

#turno
import random

turno = random.randint(1,2)
print(f"turno: jugador {turno} empieza")
jugador1 = 0
jugador2 = 0
pierde_turno1 = False
pierde_turno2 = False
#dado
while True:
  input("Presione ENTER para lanzar el dado...")
  dado = random.randint(1, 6)
  print(f"dado: {dado}")
#jugador 1 en turno
  if turno == 1:
    jugador1 += dado
    print(f"jugador 1: {jugador1}")
    if jugador1 == 5:
      print(f"Caiste en modo automatico, repite el turno inmediatamente(juaga dos veces)")
      turno == 1, input("presione enter para lanzar el dado otra vez")
      dado = random.randint(1, 6)
      print(f"Dado: {dado}")
      jugador1 += dado
      print(f"jugador 1: {jugador1}")
    elif jugador1 == 9:
      print(f"caiste en un corto circuito, retrocede al inico del dablero")
      jugador1 = 1
    elif jugador1 == 12:
      print(f"caiste en ajuste de torque. avanzas 2 casillas")
      jugador1 += 2
    elif jugador1 == 14:
      print(f"caiste en restablecimiento de energia!, avanzas 6 casillas, los demas jugadores retroceden 6")
      jugador1 += 6
      jugador2 -= 6
    elif jugador1 == 16:
      print(f"caiste en mantenimiento preventivo, todos los jugadores retroceden 1 casilla")
      jugador1 -= 1
      jugador2 -= 1
    elif jugador1 == 18:
      print(f"caiste en falla de motor, retrocedes 2 casillas")
      jugador1 -= 2
    elif jugador1 == 22:
      print(f"caiste en actualizacion de software, ir hasta la otra cailla de actualizacion")
      jugador1 = 36
    elif jugador1 == 26:
      print(f"caiste en sobrecarga de electrica , pierdes el siguiente turno turno")
      pierde_turn1 = True
    elif jugador1 == 29:
      print(f"caiste en ajuste de torque, avanzas 2 casillas")
      jugador1 += 2
    elif jugador1 == 31:
      print(f"caiste en falla mecanica, retrocede a la casilla anterior de mantenimiento")
      jugador1 == 28
    elif jugador1 == 34:
      print(f"caiste en corto circuiro, el jugador retrocede al inicio del tablero")
      jugador1 = 1
    elif jugador1 == 36:
      print(f"caiste en actualizacion de software, ir a hasta la otra casilla de actualizacion")
      jugador1 = 22
    elif jugador1 == 42:
      print(f"caiste en optamizacion del sistema, el jugador elije avanzar 3 0 retroceder 2 casillas")
      if jugador1 == input("3"):
        jugador1 += 3
      if jugador1 == input("2"):
        jugador1 -= 2
    elif jugador1 == 44:
      print(f"caiste en sensor fuera de calibracion, salta hasta la casilla de calibrar sensor")
      jugador1 = 7
      if pierde_turno1 == False:
        turno = 1
    else:
      pierde_turno1 = False
    turno = 2
  #jugador 2 en turno
  else:
    jugador2 += dado
    print(f"jugador 2: {jugador2}")
    if jugador2 == 5:
      print(f"Caiste en modo automatico, repite el turno inmediatamente(juaga dos veces)")
      turno == 2, input(f"presione enter para lanzar el dado otra vez")
      dado = random.randint(1, 6)
      print(f"Dado: {dado}")
      jugador2 += dado
      print(f"jugador 2: {jugador2}")
    elif jugador2 == 9:
      print(f"caiste en un corto circuito, retrocede al inico del dablero")
      jugador1 = 1
    elif jugador2 == 12:
      print(f"caiste en ajuste de torque. avanzas 2 casillas")
      jugador2 += 2
    elif jugador2 == 14:
      print(f"caiste en restablecimiento de energia!, avanzas 6 casillas, los demas jugadores retroceden 6")
      jugador2 += 6
      jugador1 -= 6
    elif jugador2 == 16:
      print(f"caiste en mantenimiento preventivo, todos los jugadores retroceden 1 casilla")
      jugador1 -= 1
      jugador2 -= 1
    elif jugador2 == 18:
      print(f"caiste en falla de motor, retrocedes 2 casillas")
      jugador2 -= 2
    elif jugador2 == 22:
      print(f"caiste en actualizacion de software, ir hasta la otra cailla de actualizacion")
      jugador2 = 36
    elif jugador2 == 26:
      print(f"caiste en sobrecarga de electrica , pierdes el siguiente turno turno")
      pierde_turno2 = True
    elif jugador2 == 29:
      print(f"caiste en ajuste de torque, avanzas 2 casillas")
      jugador2 += 2
    elif jugador2 == 31:
      print(f"caiste en falla mecanica, retrocede a la casilla anterior de mantenimiento")
      jugador2 == 28
    elif jugador2 == 34:
      print(f"caiste en corto circuiro, el jugador retrocede al inicio del tablero")
      jugador2 = 1
    elif jugador2 == 36:
      print(f"caiste en actualizacion de software, ir a hasta la otra casilla de actualizacion")
      jugador1 = 22
    elif jugador2 == 42:
      print(f"caiste en optamizacion del sistema, el jugador elije avanzar 3 0 retroceder 2 casillas")
      if jugador2 == input("3"):
        jugador2 += 3
      if jugador2 == input("2"):
        jugador2 -= 2
    elif jugador2 == 44:
      print(f"caiste en sensor fuera de calibracion, salta hasta la casilla de calibrar sensor")
      jugador2 = 7
    if pierde_turno2 == False:
      turno = 1
    else:
      pierde_turno2 = False

  #jugador 1 gana
  if jugador1 == 50:
          print("Ganador Jugador 1")
          break
  #jugador 1 cae fuera del tablero
  elif jugador1 > 50:
          jugador1 -= dado
          print(f"Has caido fuera del tablero regresas a la casilla {jugador1}")
  #jugador 2 gana
  if jugador2 == 50:
          print("Ganador Jugador 2")
          break
  #jugador 2 cae fuera del tablero
  elif jugador2 > 50:
          jugador2 -= dado
          print(f"Has caido fuera del tablero regresas a la casilla {jugador2}")

  print(f"Turno: Jugador {turno}")
```
